```python
#200 days moving average code 

import pandas as pd
import yfinance as yf

# Load S&P 500 data
sp500_data = yf.download('^GSPC', start='2005-01-01', end='2023-08-31')

# Calculate the 200-day moving average
sp500_data['SMA200'] = sp500_data['Close'].rolling(window=200).mean()

# Initialize variables
position = None
initial_investment = None
portfolio_value = None
positions = []

# Implement the 200-day moving average trading strategy
for i in range(200, len(sp500_data)):
    if sp500_data['Close'][i] > sp500_data['SMA200'][i] and position != 'long':
        # Buy when the close price crosses above the 200-day moving average
        position = 'long'
        positions.append(('buy', sp500_data.index[i], sp500_data['Close'][i]))
        if initial_investment is None:
            initial_investment = sp500_data['Close'][i]
            portfolio_value = initial_investment
    elif sp500_data['Close'][i] < sp500_data['SMA200'][i] and position == 'long':
        # Sell when the close price crosses below the 200-day moving average
        position = None
        positions.append(('sell', sp500_data.index[i], sp500_data['Close'][i]))
        if initial_investment is not None:
            portfolio_value = sp500_data['Close'][i]

# Calculate the percentage profit
percentage_profit = ((portfolio_value - initial_investment) / initial_investment) * 100

print(f"Initial Investment: ${initial_investment:.2f}")
print(f"Final Portfolio Value: ${portfolio_value:.2f}")
print(f"Percentage Profit: {percentage_profit:.2f}%")
```

profit percentage is alot more than buy and hold

```python
#if include transaction fee

import pandas as pd
import yfinance as yf

# Load S&P 500 data
sp500_data = yf.download('^GSPC', start='2005-01-01', end='2023-08-31')

# Calculate the 200-day moving average
sp500_data['SMA200'] = sp500_data['Close'].rolling(window=200).mean()

# Initialize variables
position = None
initial_investment = None
portfolio_value = None
positions = []
transaction_fee_percent = 0.005  # 0.5% transaction fee when buying

# Implement the 200-day moving average trading strategy
for i in range(200, len(sp500_data)):
    if sp500_data['Close'][i] > sp500_data['SMA200'][i] and position != 'long':
        # Buy when the close price crosses above the 200-day moving average
        position = 'long'
        positions.append(('buy', sp500_data.index[i], sp500_data['Close'][i]))
        if initial_investment is None:
            initial_investment = sp500_data['Close'][i]
            portfolio_value = initial_investment
        # Apply transaction fee when buying
        portfolio_value -= sp500_data['Close'][i] * transaction_fee_percent
    elif sp500_data['Close'][i] < sp500_data['SMA200'][i] and position == 'long':
        # Sell when the close price crosses below the 200-day moving average
        position = None
        positions.append(('sell', sp500_data.index[i], sp500_data['Close'][i]))
        if initial_investment is not None:
            portfolio_value = sp500_data['Close'][i]

# Calculate the percentage profit, considering transaction fees
percentage_profit = ((portfolio_value - initial_investment) / initial_investment) * 100

print(f"Initial Investment: ${initial_investment:.2f}")
print(f"Final Portfolio Value: ${portfolio_value:.2f}")
print(f"Percentage Profit (including transaction fees): {percentage_profit:.2f}%")
```

```python
import yfinance as yf
import pandas as pd
from datetime import datetime

# Load S&P 500 data up to today
end_date = datetime.today().strftime('%Y-%m-%d')
start_date = (datetime.today() - pd.DateOffset(years=5)).strftime('%Y-%m-%d')
sp500_data = yf.download('^GSPC', start=start_date, end=end_date)

# Calculate the 200-day moving average
sp500_data['SMA200'] = sp500_data['Close'].rolling(window=200).mean()

# Initialize variables
recommendations = []
actions = []

# Analyze data up to today
for i in range(200, len(sp500_data)):
    latest_close = sp500_data['Close'].iloc[i]
    sma200 = sp500_data['SMA200'].iloc[i]

    # Set a threshold for decision-making (you can adjust this)
    threshold = 0.01  # 1% threshold

    # Make a recommendation based on the 200-day moving average
    if latest_close > sma200 * (1 + threshold):
        recommendation = "Buy"
        action = latest_close  # Buy at the closing price
    elif latest_close < sma200 * (1 - threshold):
        recommendation = "Sell"
        action = latest_close  # Sell at the closing price
    else:
        recommendation = "Hold"
        action = None

    recommendations.append(recommendation)
    actions.append(action)

# Count the number of Buy, Sell, and Hold recommendations
buy_count = recommendations.count("Buy")
sell_count = recommendations.count("Sell")
hold_count = recommendations.count("Hold")

# Calculate profit based on actions
initial_balance = 10000.0  # Initial balance (e.g., $10,000)
balance = initial_balance

# Track the buy and sell actions for profit calculation
buy_actions = []
sell_actions = []

for i in range(len(recommendations)):
    if recommendations[i] == "Buy" and actions[i] is not None:
        balance -= actions[i] * 0.005  # Subtract 0.5% transaction fee for buying
        buy_actions.append(actions[i])
    elif recommendations[i] == "Sell" and actions[i] is not None:
        balance += actions[i]  # Add the selling price
        sell_actions.append(actions[i])

# Calculate total profit as a percentage of the initial balance
total_profit_percentage = ((balance - initial_balance) / initial_balance) * 100

# Print the counts, today's recommendation, and total profit percentage
print(f"Number of 'Buy' Recommendations up to today: {buy_count}")
print(f"Number of 'Sell' Recommendations up to today: {sell_count}")
print(f"Number of 'Hold' Recommendations up to today: {hold_count}")
print(f"Total Profit Percentage: {total_profit_percentage:.2f}%")

# Print the list of buy and sell actions for reference
print("Buy Actions:", buy_actions)
print("Sell Actions:", sell_actions)
```
